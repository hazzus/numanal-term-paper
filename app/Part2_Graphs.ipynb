```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
time = []
ga_cl = []
ga_cl_2 = []
ga_cl_3 = []
volume = []

with open('part2Res', 'r') as file:
    for line in file.readlines():
        t, _1, _2, _3, v = list(map(float, line.split()))
        time.append(t)
        ga_cl.append(_1)
        ga_cl_2.append(_2)
        ga_cl_3.append(_3)
        volume.append(v)

time = np.array(time)
ga_cl = np.array(ga_cl)
ga_cl_2 = np.array(ga_cl_2)
ga_cl_3 = np.array(ga_cl_3)
volume = np.array(volume)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 8))

ax.plot(1 / time, np.log(np.abs(ga_cl)), linewidth=3, label='$Ln(G_{GaCl})$')
ax.plot(1 / time, np.log(np.abs(ga_cl_2)), linewidth=3, label='$Ln(G_{GaCl_2})$')
ax.plot(1 / time, np.log(np.abs(ga_cl_3)), linewidth=3, label='$Ln(G_{GaCl_3})$')

ax.set_title('Диаграмма Аррениуса', rotation=0, pad=15, fontsize=25)
ax.set_xlabel('$\\frac{1}{Time}$', rotation=0, labelpad=20, fontsize=25)
ax.set_ylabel('$Ln(G_{GaCl_{...}})$', rotation=0, labelpad=50, fontsize=20)

ax.legend(loc=4, prop={'size': 15})

ax.minorticks_on()
ax.grid(b=True, which=r'major', linestyle='-')
ax.grid(b=True, which=r'minor', linestyle='--')

plt.savefig('graphPart2_1.png', dpi=500)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 8))

ax.plot(1 / time, np.log(np.abs(volume)), linewidth=3)

ax.set_title('Диаграмма Аррениуса', rotation=0, pad=15, fontsize=25)
ax.set_xlabel('$\\frac{1}{Time}$', rotation=0, labelpad=20, fontsize=25)
ax.set_ylabel('$Ln(V^{e_{Ga}})$', rotation=0, labelpad=50, fontsize=20)

ax.minorticks_on()
ax.grid(b=True, which=r'major', linestyle='-')
ax.grid(b=True, which=r'minor', linestyle='--')

plt.savefig('graphPart2_2.png', dpi=500)
plt.show()
```
